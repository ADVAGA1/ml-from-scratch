# Simple Linear Regression

```python
from LinearRegression import SimpleLinearRegression
```

```python
data = pd.read_csv("data/Salary_dataset_simpleLr.csv", index_col=0)
```

```python
data.head()
```

```python
X = data.YearsExperience.to_list()
y = data.Salary.to_list()
```

```python
model = SimpleLinearRegression(iterations=1000000, learning_rate=0.001)
model.fit(X, y)
```

```python
predictions = model.predict(X)
```

```python
model.eval(predictions, y)
```
